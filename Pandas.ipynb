# Making Changes to the DataFrame

```python
data.head(5)
```

```python
#Adding a column
# adding a new column called Total that is 
data['Total'] = data['HP'] + data['Attack'] + data['Defense'] + data['Sp. Atk'] + data['Sp. Def'] + data['Speed']
data.head(5)
data.sort_values('Total', ascending = False)
```

```python
#Deleting a column
# dropping the Total column from table data
data.drop(columns=['Total'])
```

```python
# Summing Multiple Columns to Create new Column
#another way to add a new column called Total that is sum of  [4:10] columns
data['Total'] = data.iloc[:,4:10].sum(axis = 1)
data.head(11)
data.sort_values('Total',ascending = False)
```

```python
# Rearranging columns
data = data[['Name','Type 1','Type 2','HP','Attack','Defense','Sp. Atk','Sp. Def','Speed','Total','Generation','Legendary']]
```

```python
# Saving our Data (CSV, Excel, TXT, etc)
data.to_csv('data_csv.csv', index = False)
data.to_excel('data_xlsx.xlsx',index = False)
```

```python
# Filtering Data (based on multiple conditions)
filterd_data = data.loc[(data['Type 1'] =='Grass' ) & (data['Type 2'] == 'Poison' ) & (data['Total'] >= 500)]
filterd_data.to_csv('filterd_data.csv', index = False)
filterd_data
```

```python
# Reset Index
filterd_data.reset_index()            # keping the old index
filterd_data.reset_index(drop = True) # droping the old index 
```

```python
# Regex Filtering (filter based on textual patterns)
data_Mega = data.loc[data['Name'].str.contains('Mega')]
# data_Mega.reset_index( drop = True)   # reseting the index and drop the old indexs
# data_Mega.sort_values('Total',ascending = False) # sorting data_Mega 
data_Mega.sort_values('Total',ascending = False).reset_index(drop = True) 
```

```python
# Conditional Changes
data.loc[data['Type 1'] == 'Flamer','Type 1'] = 'Fire'
data
```

# Aggregate Statistics using Groupby (Sum, Mean, Counting)

```python
# here we are grouping the pokemon by Type 1 to calculat a Statistical parameter
data.groupby('Type 1').mean()
```

```python
data.groupby('Type 1').sum()
```

```python
# grouping data by Type 1 then count names in each group that represents num of pokemons in each group 
data.groupby('Type 1').count()['Name'] 
```

```python
# grouping by multiple parameters
data.groupby(['Type 1','Type 2']).count()['Name'] 
```
